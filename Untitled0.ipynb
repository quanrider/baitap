```python
#cau 13
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))

    for i in range(n):
        c.append(a[i] + b[i])

    if len(a) > len(b):
        c.extend(a[n:])
    else:
        c.extend(b[n:])

    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print("Mảng kết quả:", tron_mang(a, b))
```
